```python
import sqlite3
import pandas as pd
```

```python
con=sqlite3.connect('sakila.db')
```

```python
sql_query='''SELECT * FROM customer '''
df=pd.read_sql(sql_query,con)
```

```python
df.head()
```

```python
# to set a function as SQL to pandas
def sql_to_df(x):
    df=pd.read_sql(x,con)
    return df
```

```python
query=''' SELECT first_name,last_name FROM customer;'''
# call the function
sql_to_df(query).head()
```

```python
query='''SELECT DISTINCT country_id FROM city'''
sql_to_df(query).head()
```

```python
query='''SELECT * FROM customer WHERE first_name='MARY';'''
sql_to_df(query).head()
```

```python
query='''SELECT * FROM customer WHERE first_name LIKE 'M%';'''
sql_to_df(query).head()
```

```python
query='''SELECT * FROM customer WHERE last_name LIKE '_ing';'''
sql_to_df(query).head()
```

```python
query='''SELECT * FROM customer WHERE last_name LIKE '%ing';'''
sql_to_df(query).head()
```

```python
query='''SELECT * FROM customer WHERE first_name GLOB '[AC]*';'''
sql_to_df(query).head()
```

```python
query='''SELECT * FROM customer ORDER BY first_name DESC;'''
sql_to_df(query).head()
```

```python
query='''SELECT COUNT (store_id) FROM customer;'''
sql_to_df(query).head()
```

```python
query='''SELECT store_id, COUNT(customer_id) FROM customer GROUP BY store_id;'''
sql_to_df(query).head()
```
